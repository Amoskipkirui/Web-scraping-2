```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd

url = 'https://books.toscrape.com/catalogue/category/books/poetry_23/index.html'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python

titles = []
prices = []
star_ratings = []
availability = []

# Dictionary to map star rating strings to integers
star_dict = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}

for article in soup.find_all('article', class_= 'product_pod'):# container
    titles.append(article.h3.a.attrs['title']) #chain to title
    prices.append(article.select('.price_color')[0].get_text()[1:])# leaves behind currency symbol
    regex = re.compile("star-rating (.*)") #
    for p in article.find_all('p', class_ = regex):
        # Extract the last value of the 'class' attribute, map to integer using the star_dict, and append to star_ratings list
        star_ratings.append(star_dict[p.attrs['class'][-1]])
    availability.append(article.select('.instock.availability')[0].get_text().strip())

df = pd.DataFrame({'Title': titles, 'Price': prices, 'Rating': star_ratings, 'Availability': availability})
df
```
